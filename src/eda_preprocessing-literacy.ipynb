```python
import matplotlib.pyplot as plt

# primary, secondary, and tertiary completion
labels = ['Primary Completion', 'Secondary Completion', 'Tertiary Enrollment']

# completion rates for primary, secondary, and tertiary
female_completion_indicators = [
    'Primary completion rate, female (% of relevant age group)',
    'Lower secondary completion rate, female (% of relevant age group)',
    'School enrollment, tertiary, female (% gross)'
]

male_completion_indicators = [
    'Primary completion rate, male (% of relevant age group)',
    'Lower secondary completion rate, male (% of relevant age group)',
    'School enrollment, tertiary, male (% gross)'
]

# Filter the data
df_female_completion = df_female[df_female['Indicator Name'].isin(female_completion_indicators)]
df_male_completion = df_male[df_male['Indicator Name'].isin(male_completion_indicators)]


years = ['1971', '1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979',
       '1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988',
       '1989', '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997',
       '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006',
       '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015',
       '2016', '2017', '2018', '2019', '2020'] 

# average values for the selected years for each indicator
female_data = df_female_completion[years].mean(axis=1).values[:3]  
male_data = df_male_completion[years].mean(axis=1).values[:3]      

# shapes shape
print("Female data shape:", female_data.shape)
print("Male data shape:", male_data.shape)

# Plot the data
x = range(len(labels))  

# bar chart
plt.bar(x, female_data, width=0.4, label='Female', align='center')
plt.bar(x, male_data, width=0.4, label='Male', align='edge')

# labels and title
plt.xlabel('Education Level')
plt.ylabel('Completion Rate (%)')
plt.title('Completion Rate of Primary, Secondary, and Tertiary Education (Female vs Male)')
plt.xticks(x, labels)
plt.legend()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


# Filter the dataset for the GPI indicator for primary school enrollment
gpi_data = df_female[df_female['Indicator Name'] == 'School enrollment, primary (gross), gender parity index (GPI)']

# Melt the dataframe to make it suitable for time series plotting
gpi_melted = gpi_data.melt(id_vars=['Country Name'], value_vars=[str(year) for year in range(1971, 2021)],
                           var_name='Year', value_name='GPI')

# Convert the Year column to numeric
gpi_melted['Year'] = pd.to_numeric(gpi_melted['Year'])

# Plot the Gender Parity Index over time for each country
plt.figure(figsize=(14, 8))  # Increase figure size

# Plot each country's GPI over time
for country in gpi_melted['Country Name'].unique():
    country_data = gpi_melted[gpi_melted['Country Name'] == country]
    plt.plot(country_data['Year'], country_data['GPI'], marker='o', label=country)

# Annotating major gender equality events
# I used 5 (I don't know what else should I use)
plt.axvline(x=1979, color='red', linestyle='--', label="1979: CEDAW Adopted")
plt.axvline(x=1995, color='green', linestyle='--', label="1995: Beijing Platform for Action")
plt.axvline(x=2000, color='blue', linestyle='--', label="2000: Millennium Development Goals")
plt.axvline(x=2010, color='purple', linestyle='--', label="2010: UN Women Founded")
plt.axvline(x=2015, color='orange', linestyle='--', label="2015: SDGs Introduced")

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Gender Parity Index (GPI)')
plt.title('Gender Parity Index Over Time for Primary School Enrollment')

# Move legend outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), title='Country')

# Apply tight layout with padding
plt.tight_layout(pad=2.0)

plt.show()
```

```python

# Compute mean GPI for years before and after 1995 (Beijing Platform)
before_1995 = gpi_melted[gpi_melted['Year'] < 1995].groupby('Country Name')['GPI'].mean()
after_1995 = gpi_melted[gpi_melted['Year'] >= 1995].groupby('Country Name')['GPI'].mean()

# Create a new dataframe for before and after comparison
df_before_after = pd.DataFrame({'Before 1995': before_1995, 'After 1995': after_1995}).reset_index()

# Melt the dataframe for easy plotting
df_melted = df_before_after.melt(id_vars='Country Name', var_name='Period', value_name='GPI')

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Country Name', y='GPI', hue='Period', data=df_melted)
plt.xticks(rotation=90)
plt.title('Comparison of GPI Before and After 1995 (Beijing Platform for Action)')
plt.tight_layout()
plt.show()
```

```python
# Correct use of pivot with keyword arguments
heatmap_data = gpi_melted.pivot(index='Country Name', columns='Year', values='GPI')

# Plot the heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=0.5, annot=False)
plt.title('GPI Across Countries Over Time')
plt.show()
```

```python

# Set a color palette
sns.set_palette("husl")  # Husl palette for distinguishable lines

# Increase figure size for better readability
plt.figure(figsize=(14, 8))

# Plot GPI for each country with distinct colors and markers
for country in gpi_melted['Country Name'].unique():
    country_data = gpi_melted[gpi_melted['Country Name'] == country]
    plt.plot(country_data['Year'], country_data['GPI'], marker='o', label=country, linewidth=2, markersize=6)

# Add shaded regions for events with appropriate transparency
plt.axvspan(1979, 1995, color='red', alpha=0.1, label='After 1979: CEDAW')
plt.axvspan(1995, 2000, color='green', alpha=0.1, label='After 1995: Beijing Platform')
plt.axvspan(2000, 2010, color='blue', alpha=0.1, label='After 2000: MDGs')
plt.axvspan(2010, 2015, color='purple', alpha=0.1, label='After 2010: UN Women')
plt.axvspan(2015, 2020, color='orange', alpha=0.1, label='After 2015: SDGs')

# Add grid for better trend visibility
plt.grid(True, linestyle='--', alpha=0.7)

# Increase font size of labels and title
plt.xlabel('Year', fontsize=14)
plt.ylabel('Gender Parity Index (GPI)', fontsize=14)
plt.title('GPI Over Time with Shaded Regions for Major Events', fontsize=16)

# Place legend outside the plot and adjust for readability
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=10, title='Country', ncol=2)

# Apply tight layout to avoid overlapping elements
plt.tight_layout()

# Show the plot
plt.show()

```
