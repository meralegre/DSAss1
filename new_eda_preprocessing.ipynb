```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


# Filter the dataset for the GPI indicator for primary school enrollment
gpi_data = df_female[df_female['Indicator Name'] == 'School enrollment, primary (gross), gender parity index (GPI)']

# Melt the dataframe to make it suitable for time series plotting
gpi_melted = gpi_data.melt(id_vars=['Country Name'], value_vars=[str(year) for year in range(1971, 2021)],
                           var_name='Year', value_name='GPI')

# Convert the Year column to numeric
gpi_melted['Year'] = pd.to_numeric(gpi_melted['Year'])

# Plot the Gender Parity Index over time for each country
plt.figure(figsize=(14, 8))  # Increase figure size

# Plot each country's GPI over time
for country in gpi_melted['Country Name'].unique():
    country_data = gpi_melted[gpi_melted['Country Name'] == country]
    plt.plot(country_data['Year'], country_data['GPI'], marker='o', label=country)

# Annotating major gender equality events
# I used 5 (I don't know what else should I use)
plt.axvline(x=1979, color='red', linestyle='--', label="1979: CEDAW Adopted")
plt.axvline(x=1995, color='green', linestyle='--', label="1995: Beijing Platform for Action")
plt.axvline(x=2000, color='blue', linestyle='--', label="2000: Millennium Development Goals")
plt.axvline(x=2010, color='purple', linestyle='--', label="2010: UN Women Founded")
plt.axvline(x=2015, color='orange', linestyle='--', label="2015: SDGs Introduced")

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Gender Parity Index (GPI)')
plt.title('Gender Parity Index Over Time for Primary School Enrollment')

# Move legend outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), title='Country')

# Apply tight layout with padding
plt.tight_layout(pad=2.0)

plt.show()
```

```python

# Compute mean GPI for years before and after 1995 (Beijing Platform)
before_1995 = gpi_melted[gpi_melted['Year'] < 1995].groupby('Country Name')['GPI'].mean()
after_1995 = gpi_melted[gpi_melted['Year'] >= 1995].groupby('Country Name')['GPI'].mean()

# Create a new dataframe for before and after comparison
df_before_after = pd.DataFrame({'Before 1995': before_1995, 'After 1995': after_1995}).reset_index()

# Melt the dataframe for easy plotting
df_melted = df_before_after.melt(id_vars='Country Name', var_name='Period', value_name='GPI')

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Country Name', y='GPI', hue='Period', data=df_melted)
plt.xticks(rotation=90)
plt.title('Comparison of GPI Before and After 1995 (Beijing Platform for Action)')
plt.tight_layout()
plt.show()
```

```python
# Correct use of pivot with keyword arguments
heatmap_data = gpi_melted.pivot(index='Country Name', columns='Year', values='GPI')

# Plot the heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=0.5, annot=False)
plt.title('GPI Across Countries Over Time')
plt.show()
```

```python


# Load the datasets
df_filtered = pd.read_csv('/Users/bognarlili/Desktop/df_filtered.csv')
prediction = pd.read_csv('/Users/bognarlili/Desktop/prediction.csv')

# Drop duplicates based on Year, Country, and Keyword Count
prediction_cleaned = prediction.drop_duplicates(subset=['Year', 'ISO-alpha3 Code', 'Keyword Count'])

# Sum the keyword mentions for each year after removing duplicates
mentions_per_year = prediction_cleaned.groupby('Year')['Keyword Count'].sum().reset_index()

# Filter out years beyond 2020
mentions_per_year = mentions_per_year[mentions_per_year['Year'] <= 2020]

# Melt the df_filtered dataset to have years as rows, filtering for years <= 2020
education_indicators = df_filtered.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name'], 
                                        value_vars=[str(year) for year in range(1971, 2021)],  # Cap at 2020
                                        var_name='Year', 
                                        value_name='Value')

# Convert 'Year' to numeric for consistency and filter out years beyond 2020
education_indicators['Year'] = pd.to_numeric(education_indicators['Year'], errors='coerce')
education_indicators = education_indicators[education_indicators['Year'] <= 2020]

# Merge with mentions data on 'Year'
merged_data = pd.merge(education_indicators, mentions_per_year, on='Year', how='inner')

# Filter for relevant indicators (enrollment and completion rates, female)
female_education_indicators = merged_data[merged_data['Indicator Name'].str.contains('enrollment|completion', case=False) & 
                                          merged_data['Indicator Name'].str.contains('female', case=False)].copy()

# Drop rows where 'Value' is NaN
female_education_indicators.dropna(subset=['Value'], inplace=True)

# Group by year and calculate the average for each indicator
average_enrollment_completion_per_year = female_education_indicators.groupby(['Year', 'Indicator Name'])['Value'].mean().unstack()

# Plot the trends over time for each indicator (without markers)
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot the enrollment and completion data on the first y-axis
for column in average_enrollment_completion_per_year.columns:
    ax1.plot(average_enrollment_completion_per_year.index, average_enrollment_completion_per_year[column], label=column)

ax1.set_xlabel('Year')
ax1.set_ylabel('Average Enrollment/Completion (% Gross)')
ax1.grid(True)

# Create a second y-axis for mentions (with black line and no markers)
ax2 = ax1.twinx()
ax2.plot(mentions_per_year['Year'], mentions_per_year['Keyword Count'], color='black', linestyle='-', label='Mentions of Gender Equality')
ax2.set_ylabel('Total Mentions of Gender Equality')

# Adjust the layout to fix overlapping elements
fig.tight_layout(pad=3.0)  # Increase padding to prevent overlap

# Place the legends outside the plot with better positioning
ax1_legend = ax1.legend(loc='upper left', bbox_to_anchor=(1.05, 1), borderaxespad=0.)  # Move legend outside the plot to the right
ax2_legend = ax2.legend(loc='upper left', bbox_to_anchor=(1.05, 0.75), borderaxespad=0.)  # Move the second legend lower

# Add titles and labels
plt.title('Trends in Female Enrollment/Completion Rates and Mentions of Gender Equality Over Time (Capped at 2020)')

# Save the plot as a PNG file
plt.savefig('enrollment_completion_mentions_trends_capped_2020.png')

# Show the plot
plt.show()

```
